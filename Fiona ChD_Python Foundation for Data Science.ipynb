3. Buatlah program untuk menghitung BMI!

```python
# Program kalkulator BMI

nama = input("Nama Anda: ")

berat = int(input("Berat anda dalam kilogram: "))
tinggi = int(input("Tinggi anda dalam centimeter: "))

BMI = (berat)/(tinggi/100)**2

print(BMI)

if BMI>0:
    if(BMI < 18.5):
        print(nama + ", Berat Anda dalam kategori underweight.")
    elif(18.5 <= BMI <= 24.9):
        print(nama + ", Berat Anda dalam kategori normal.")
    elif(BMI<= 29.9):
        print(nama + ", Berat Anda dalam kategori overweight.")
    elif(BMI<=34.9):
        print(nama + ", Berat Anda dalam kategori obesity.")
else:
    print("Maaf, kategori berat Anda tidak terdefinisi")
```

5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5

```python
def status_bmi(bmi):
  if bmi < 18.5:
    return 'Underweight'
  elif 18.5 <= bmi <= 24.9:
    return 'Normal'
  elif bmi <= 29.9:
    return 'Overweight'
  elif bmi <= 34.9:
    return 'Obesity'
  else:
    return 'Undefined'
```

```python
df_10['Status BMI'] = df_10['BMI'].apply(status_bmi)
```

#**Soal 2**

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:

Input
>> Money Capital : 100000
>> Investment Tenure (years) : 2

Output
>> Year 1, your total investment : 125000.0
>> Year 2, your total investment : 156250.0

```python
def calculate_investment(modal, tenure):
    # Cek apakah modal dan tenure berada dalam rentang yang valid
    if not 0 < modal <= 1000000000:
        print("Error: Modal investasi harus berada dalam rentang antara 1 hingga 1.000.000.000.")
        return
    if not 0 < tenure <= 25:
        print("Error: Jangka waktu investasi harus berada dalam rentang antara 1 hingga 25 tahun.")
        return

    # Hitung keuntungan tahunan (25%)
    keuntungan_tahunan = modal * 0.25

    # Iterasi untuk setiap tahun
    for year in range(1, tenure + 1):
        # Hitung total investasi untuk tahun tersebut
        total_investasi = modal + keuntungan_tahunan * (year - 1)
        print(f"Year {year}, your total investment: {total_investasi}")

# Input modal investasi dan jangka waktu investasi dari pengguna
modal = int(input("Money Capital: "))
tenure = int(input("Investment Tenure (years): "))

# Hitung dan tampilkan informasi akumulasi investasi
calculate_investment(modal, tenure)
```
